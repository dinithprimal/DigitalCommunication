```python
import matplotlib.pyplot as plt
import math
import numpy as np

pi = 22/7;

fs=10;     #sampling frequency
t1=0.5;
#time base - upto 1 second
t = np.arange(0, t1, 0.001);
#print(t)

amplitude = np.sin(2*pi*fs*t + (1/3*pi));

plt.plot(t, amplitude)

plt.title('Sine wave')

# Give x axis label for the sine wave plot
plt.xlabel('Time')

# Give y axis label for the sine wave plot
plt.ylabel('Amplitude')

plt.grid(True, which='both')

plt.axhline(y=0, color='k')

plt.show()

# Display the sine wave
plt.show()
```

```python
amplitude_Squre = np.sign(np.sin(2*pi*fs*t))

plt.plot(t, amplitude_Squre)

plt.title('Square wave')

plt.xlabel('Time')

# Give y axis label for the sine wave plot
plt.ylabel('Amplitude')

plt.grid(True, which='both')

plt.axhline(y=0, color='k')

```

```python
sigma = 0.1;

t1 = np.arange(-0.5, 0.5, 0.001);

gausian = (1/((2*pi)**0.5)*sigma)*np.exp(-(((t1)**2)/(2*(sigma**2))));

plt.plot(t1, gausian)

plt.title('Gausian Distribution')

plt.xlabel('Time')

# Give y axis label for the sine wave plot
plt.ylabel('Amplitude')

plt.grid(True, which='both')

plt.axhline(y=0, color='k')
```
