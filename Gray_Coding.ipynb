```python
# This method is used 2 function for store the memory for both decimal to binary values
# and gray codeing. Then using functions get the gray for bitstream

import matplotlib.pyplot as plt
import math
import random
import numpy as np

###############################################################################
# Function for Convert the decimal numbers to binary
def DecimalToBinary(num):
     
    if num > 1:
        val = DecimalToBinary(num // 2)
        num = str(val)+""+str(num % 2)
    return ""+str(num)


###############################################################################
# Funciotn for Generate the gray coding
def grayCoding(arr_dec2bin,k):

  grayArr = []
  for i in range(0,2**k):
    temp = ""
    for j in range(0,len(arr_dec2bin[i])):
      if(j == 0):
        temp = temp + arr_dec2bin[i][j]
      else:
        temp = temp + str(int(arr_dec2bin[i][j-1]) ^ int(arr_dec2bin[i][j]))

    grayArr.append(temp)

  return grayArr


###############################################################################
# Mapping the binary symbols to Gray code sysmbols
def mapGrayCoding(bitStream,k_value):

  # Mapiing decimal to binary
  arr_dec2bin = []
  for i in range(0,2**k_value):
    val = DecimalToBinary(i)
    if (len(val)!=k_value):
      while len(val) != k_value:
        val = "0"+val
   
    arr_dec2bin.append(val)

   #arr_dec2bin.append(DecimalToBinary(i))

  # Print noraml binary codes
  print("Normal Codes for Amplitude Levels")

  for k in range(0,2**k_value):
    print("Level ",k," - ",arr_dec2bin[k])

  grayArr = grayCoding(arr_dec2bin,k_value)

  # Print Gray Codes
  print("\n\nGray Codes for Amplitude Levels")

  for k in range(0,2**k_value):
    print("Level ",k," - ",grayArr[k])

  outBitSequence = []
  j = 1
  n = 0
  temp = ""
  while(n < len(bitStream)):
    if(j<=k_value):
      temp = temp + str(bitStream[n])
      n = n + 1
      j = j + 1
      if(n == (len(bitStream))):
        sym = grayArr[arr_dec2bin.index(temp)]
        for l in range(0,len(sym)):
          outBitSequence.append(int(sym[l]))

    else:
      j = 1
      
      sym = grayArr[arr_dec2bin.index(temp)]
      
      temp = ""
      for l in range(0,len(sym)):
        outBitSequence.append(int(sym[l]))

  return outBitSequence


###############################################################################

# Number of levels in Quantization
# M-PAM -----> M = 8
PAM = 8

# Get the value of a symbol length
k = int(math.log(8,2))

# Generating a random bit sequence
print("Input Bit Stream - ", end="")
bitSquence = [random.randint(0, 1) for _ in range(k) for _ in range(8)]

print(bitSquence)
print("")


# Call the mapping function



outBitSequence = mapGrayCoding(bitSquence,k)

print("\n\nInput Bit Stream  - ", end="")
print(bitSquence)
print("\n\nOutput Bit Stream - ", end="")
print(outBitSequence)
print("")
```
